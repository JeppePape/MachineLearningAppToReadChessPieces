## L03/Supergruppe_diskussion.ipynb: Resume


### Resume af supergruppe diskussion

#### _Look at the Big Picture_

The big picture is very much the "domain" view of the task at hand. Most, if not all, of the necessary domain knowledge should be found and disseminated to the machine learning team, from the "product owner" - ie. the relevant stakeholders. This includes checking for assumptions made by those stakeholders, including meaning embedded in specialized terminology. There are two main takeaways the team should take from this step:

* An understanding of the biases, cutoffs and other details of the input data.
* An understanding of the type and precision needed for the output data.

To test that there is a common understanding and "language" between the machine learning team and the product owner, it is a good idea that the former frames the problem with an eye to the entire domain. That is, delivers a summary that describes how the "product" will fit into the existing business. This should of course be approved by the product owner.

Finally, a performance measure must be set and agreed to.


#### _Get the Data_

This step is - almost - as simple as just downloading the data. This step should be automated, so future datasets can be downloaded automatically and added to the product pipeline easily. The more incremental the data, the more important a smooth pipeline in the system.

This is also the time to check that all assumptions have been found. A quick look at the data may reveal cutoff points, outliers or noise that make using some subset of the data useless or problematic. 

Additionally, test data is set aside in this step.


#### _Explore and Visualize the Data to Gain Insights_ 

It may be possible to learn more about the usefulness, biases, and problems in the dataset by a more thorough study. One tool is simple visualization, where plotting various features may show certain patterns or outliers. These may point to problems - such as a cutoff value that gives an "false clustering". Another tool is to check the correlations between combinations of features; other patterns worth investigating may appear. For instance, some features have no correlations with any other, and not be worth including.

It is also possible to check certain assumptions by checking the relevant features against each other in depth, using these methods. For example, in a dataset describing various systems, the weight of a some component may correlate 1-to-1 to the total weight, or be a simple constant, and not be worth including.


#### _Prepare the Data for Machine Learning Algorithms_

If there is missing data, the relevant feature or samples must be either be dropped, or padded using something like a mean or median value. Some data may also need interpretation, such as a string of text that describe a category, before being useful. It may be necessary to go back to the domain view of the problem to decide whether dat should be dropped, how to interpret categories, etc.

Once the dataset is complete and contains only numerical data, the feature values should be scaled and normalized, as most models work best when all values fall within the same range.

To ease the workflow, and allow easy replication, some form of pipelines should be implemented for all the above task. Thus, when new data is available, or the test data is to be used, the same transformations can be easily done or even automated.


#### _Select and Train a Model_

Here, we train an actual model on the actual data. This may include trying out various models on the data, to find the best one for the data. This is not just a question of the best fit, but also requires looking for for over-fitting. When suspecting overfitting, we may want to use cross-validation; train the model against one subset (a number of "folds") and test it against another subset (often a single "fold").


#### _Fine-Tune Your Model_

Finetuning the chosen model (or a small subset of candidate models found) consists of finetuning their hyperparameters. This step should also be automated, using either: A grid search that tests a combinations of a chosen subset of different hyperparameter inputs. Or, a randomized search which performs a given number of tests (iterations) with randomized variations on the hyperparameters.

Some data preparation steps, such as what features to include, whether to pad out missing data, etc. can be included as a kind of hyperparameters in this step.

Finally, when fine-tuning yields little further gains, the model can be tested against the orignally set-aside test data.


#### _Launch, Monitor, and Maintain Your System_

If the final test is passed, the system can be launched. In this phase, the system should still be monitored and adjusted if the model no longer performs as it did during development. This may because of changes to the input data, the domain has somehow changes (local economic changes, new schools opening, and similar could all change something like predictions of the housing market). For some changes, re-training the model with more recent data (and deliberately not including outdated data) may be enough. Other changes may be more systemic and have changed the enture domain "system", such as macro-economic changes, housing trends, etc.

### Beskrivelse af Dataanalyse

a) Our O4 project will be about training a machine learning algorithm to recognize chess positions from small, computer-generated images. The most optimistic goal would be an output of a FEN-string; the most commonly used short, formal description of a given chess position. We will also define other, more simple parameters for partial success; such as correctly guessing the correct number of pieces, which board fields contain a piece, and similar. We will not consider the rules of chess and whether a given position is legal; only whether the position in the image is properly described.

b) Our dataset is found on Kaggle, here:
https://www.kaggle.com/datasets/koryakinp/chess-positions?resource=download
It consists of 100000 images of chess positions, in various colors and using a handful of different sets of graphics for the pieces. This provides some variety, and hope that the trained algorithm will not become overfitted to a specific set of graphics - although this is still a posible concern.

c) Our dataset is made up of 100000 jpeg files, each containing a 400x400-pixel image of a chess position, and named for the corresponding FEN string, for instance:
1B1B2K1-1B6-5N2-6k1-8-8-8-4nq2.jpeg
Every pixel will be a feature, for a total of 160,000 features per sample.
Each individual feature will likely need to be downsampled: We have no use for the full range of possible colors in a jpeg image, and the nature of the images is such that it is unlikely that two colors that are different in the original jpeg will be collapsed into one in the resulting, downsampled version.


* d) The algorithm trained on the dataset will hopefully be able to correctly analyse similar images, and give at least some information about the chess position in it. Our tests will initially be run on a set-aside sample of images from the original dataset, but it would be interesting to test it on other computer generated chess position images; we may find that there are enough similarities between how a given piece is represented in various chess sets, that the algorithm is able to correctly identify positions in a previously unseen chess set.

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
from glob import glob
import concurrent.futures
import multiprocessing
from tqdm import tqdm
import random

# Function to apply variance and group colors within a given tolerance
def is_within_tolerance(color1, color2, tolerance=4):
    """Check if two colors are within the given RGB tolerance."""
    return all(abs(c1 - c2) <= tolerance for c1, c2 in zip(color1, color2))

# Function to quantize colors (reduce sensitivity to small variations)
def quantize_color(color, factor=30):
    return tuple((color // factor) * factor)

# Function to calculate the dominant color by finding the most frequent pixel value
def get_dominant_color(image):
    pixels = image.reshape(-1, 3)
    dominant_color = np.mean(pixels, axis=0)  # Averaging pixels for faster result
    return tuple(dominant_color.astype(int))  # Convert NumPy array to tuple (fixes error)

# Function to load and resize images for consistent processing
def load_image(image_path, size=(50, 50)):  # Smaller size for faster processing
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return cv2.resize(image, size)

# Function to detect image type: grayscale, RGB, etc.
def detect_image_type(image_path):
    img = cv2.imread(image_path)
    if len(img.shape) == 2:
        return 'Grayscale'
    elif len(img.shape) == 3 and img.shape[2] == 3:
        return 'RGB'
    elif len(img.shape) == 3 and img.shape[2] == 4:
        return 'Multispectral'
    return 'Unknown'

# Function to detect image dimensions, aspect ratio, and size in bytes
def get_image_properties(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        height, width, channels = img.shape
        image_size = os.path.getsize(image_path)  # Size in bytes
        aspect_ratio = round(width / height, 2)  # Rounded aspect ratio to 2 decimal places
        return (width, height, aspect_ratio, image_size)
    return (0, 0, 0, 0)

# Function to process each image to extract properties, including dominant color
def process_image(image_path):
    filename = os.path.basename(image_path)
    
    # Load and resize the image
    image = load_image(image_path)
    
    # Calculate the dominant color (faster using averaging)
    dominant_color = get_dominant_color(image)
    
    # Detect image properties: dimensions, aspect ratio, and size
    width, height, aspect_ratio, image_size = get_image_properties(image_path)
    
    # Detect image type
    img_type = detect_image_type(image_path)
    
    # Get file type
    file_type = os.path.splitext(filename)[-1].lower()
    
    return {
        'filename': filename,
        'img_type': img_type,
        'file_type': file_type,
        'dominant_color': dominant_color,  # Now stored as a tuple
        'width': width,
        'height': height,
        'aspect_ratio': aspect_ratio,
        'image_size': image_size,
        'image_path': image_path  # Include the path for later use in display
    }

# Function to group colors with variance tolerance
def group_colors_with_tolerance(dominant_color_counts, tolerance=4):
    grouped_colors = {}
    
    for color, count in dominant_color_counts.items():
        matched = False
        # Try to find a group that matches the color within tolerance
        for grouped_color in grouped_colors:
            if is_within_tolerance(color, grouped_color, tolerance):
                grouped_colors[grouped_color] += count
                matched = True
                break
        
        # If no group was found, add this color as a new group
        if not matched:
            grouped_colors[color] = count
    
    return grouped_colors

# Function to process images in parallel using a thread pool with progress bar
def process_images_in_parallel(image_files, max_workers=None):
    results = []
    
    # Wrap the image_files iterable with tqdm to display progress bar
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_image = {executor.submit(process_image, image_path): image_path for image_path in image_files}
        
        # tqdm progress bar for each completed image
        for future in tqdm(concurrent.futures.as_completed(future_to_image), total=len(image_files), desc="Processing Images"):
            results.append(future.result())
    
    return results

# Function to display images from 10 different color groups
def display_sample_images_by_color(results, n_samples=10):
    # Group images by their dominant color
    color_groups = defaultdict(list)
    for item in results:
        color_groups[item['dominant_color']].append(item['image_path'])
    
    # Select 10 random dominant colors
    selected_colors = random.sample(list(color_groups.keys()), min(n_samples, len(color_groups)))
    
    # Display one image from each selected color group
    fig, axes = plt.subplots(1, n_samples, figsize=(25, 15))  # Larger image size
    for i, color in enumerate(selected_colors):
        img_path = random.choice(color_groups[color])  # Choose one random image from the group
        img = load_image(img_path, size=(300, 300))  # Larger image size for display
        axes[i].imshow(img)
        axes[i].axis('off')
        axes[i].set_title(f"Color: {color}")
    
    plt.show()

# Function to summarize the results
def summarize_results(results):
    total_images = len(results)
    total_size = sum(item['image_size'] for item in results)
    
    # Count image formats and file types
    format_file_counts = Counter((item['img_type'], item['file_type']) for item in results)
    
    # Count dominant colors
    dominant_color_counts = Counter(item['dominant_color'] for item in results)
    
    # Group dominant colors with tolerance
    grouped_colors = group_colors_with_tolerance(dominant_color_counts, tolerance=4)
    
    # Count image dimensions (width x height)
    dimension_counts = Counter((item['width'], item['height']) for item in results)
    
    # Count aspect ratios
    aspect_ratio_counts = Counter(item['aspect_ratio'] for item in results)

    # Print the summary
    print(f"Total images: {total_images}")
    print(f"Total dataset size: {total_size / (1024 * 1024):.2f} MB")
    
    print("\nImage Format and File Type Counts (e.g., RGB, JPEG):")
    for (img_type, file_type), count in format_file_counts.items():
        print(f"{count} images, {img_type}, {file_type}")
    
    print(f"\nNumber of unique dominant colors found: {len(grouped_colors)}")
    
    print("\nTop 5 Dominant Color Counts (with tolerance):")
    for color, count in Counter(grouped_colors).most_common(5):  # Show only top 5 dominant colors
        print(f"Color {color}: {count} images")
    
    print("\nImage Dimension Counts (Width x Height):")
    for (width, height), count in dimension_counts.items():
        print(f"{count} images with dimensions {width} x {height}")
    
    print("\nAspect Ratio Counts:")
    for aspect_ratio, count in aspect_ratio_counts.items():
        print(f"{count} images with aspect ratio {aspect_ratio}")
    
    # Display 10 images from 10 different color dominant groups
    display_sample_images_by_color(results)

# Main function to run the optimized process
def main():
    # Get the number of CPU cores (16 threads in your case)
    num_workers = multiprocessing.cpu_count()
    
    # Load the .jpeg images
    image_folder = './dataset/combined/'  # Set this to your dataset path
    image_files = glob(os.path.join(image_folder, '*.jpeg'))
    
    # Process the images in parallel using limited threads
    results = process_images_in_parallel(image_files, max_workers=int(num_workers * 0.75))  # Use 75% CPU
    
    # Summarize the results and display sample images
    summarize_results(results)

if __name__ == "__main__":
    main()
```

So as we can see there is a total of 100.000 picture in the dataset, all jpegs, all rgb and all 400 x 400 dimension. Alongside this we calculated an estimated dominant color count, with a tolerance of +/- 5 (since the different amount of pieces and their colors can have an overall big impact on this meassurement, leading identical boards + pieces to be recognized as different "types", unless we give it a slight tolerance). Even with such a tolerance the exact number of different combinations of board colors and piece colors in the set isn´t 100% accurate however it is a good estimate of the amount.

```python
%matplotlib inline

import sys
import pickle
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def LoadDataFromL01():
    filename = "itmal_l01_data.pkl"
    with open(f"{filename}", "rb") as f:
        (X, y) = pickle.load(f)
        return X, y

X, y = LoadDataFromL01()

print(f"X.shape={X.shape},  y.shape={y.shape}")

assert X.shape[0] == y.shape[0]
assert X.ndim == 2
assert y.ndim == 1  # did a y.ravel() before saving to picke file
assert X.shape[0] == 29

# re-create plot data (not stored in the Pickel file)
m = np.linspace(0, 60000, 1000)
M = np.empty([m.shape[0], 1])
M[:, 0] = m

print("OK")
```

Plot models as defined from task description.

```python
def isNumpyData(t: np.ndarray, expected_ndim: int):
    assert isinstance(expected_ndim, int), f"input parameter 'expected_ndim' is not an integer but a '{type(expected_ndim)}'"
    assert expected_ndim>=0, f"expected input parameter 'expected_ndim' to be >=0, got {expected_ndim}"
    if t is None:
        print("input parameter 't' is None", file=sys.stderr)
        return False
    if not isinstance(t, np.ndarray):
        print("excepted numpy.ndarray got type '{type(t)}'", file=sys.stderr)
        return False
    if not t.ndim==expected_ndim:
        print("expected ndim={expected_ndim} but found {t.ndim}", file=sys.stderr)
        return False
    return True

def PlotModels(model1, model2, X: np.ndarray, y: np.ndarray, name_model1: str, name_model2: str):
    
    # NOTE: local function is such a nifty feature of Python!
    def CalcPredAndScore(model, X: np.ndarray, y: np.ndarray,):
        assert isNumpyData(X, 2) and isNumpyData(y, 1) and X.shape[0]==y.shape[0]
        y_pred_model = model.predict(X)
        score_model = r2_score(y, y_pred_model) # call r2
        return y_pred_model, score_model    

    assert isinstance(name_model1, str) and isinstance(name_model2, str)

    y_pred_model1, score_model1 = CalcPredAndScore(model1, X, y)
    y_pred_model2, score_model2 = CalcPredAndScore(model2, X, y)

    plt.plot(X, y_pred_model1, "r.-")
    plt.plot(X, y_pred_model2, "kx-")
    plt.scatter(X, y)
    plt.xlabel("GDP per capita")
    plt.ylabel("Life satisfaction")
    plt.legend([name_model1, name_model2, "X OECD data"])

    l = max(len(name_model1), len(name_model2))
    
    print(f"{(name_model1).rjust(l)}.score(X, y)={score_model1:0.2f}")
    print(f"{(name_model2).rjust(l)}.score(X, y)={score_model2:0.2f}")
```

A mlp and linear regrssion models are defined fitted and predicted results are plotted

```python
mlp = MLPRegressor(hidden_layer_sizes=(10, ),
                   solver='adam',
                   activation='relu',
                   tol=1E-5,
                   max_iter=100000,
                   verbose=False)
linreg = LinearRegression()

mlp.fit(X, y)
linreg.fit(X, y)

PlotModels(linreg, mlp, X, y, "lin.reg", "MLP")
```

Plotting function used

```python
def myXYPlot(X, y_true, y_predict):
    score = r2_score(y_true,y_predict)
    print(f"score = {score:0.2f}")
    plt.plot(X, y_predict,"r.-", label='Predicted')
    plt.scatter(X, y_true, label='True')
    plt.xlabel('GDP per capita')
    plt.ylabel('Life satisfaction')
    plt.legend()
```

### Qa) Create a Min/max scaler for the MLP

The data input matrix, X, and target vector, y, are manually scaled to a range of [0;1].
Afterwards both models are retrained and new predictions are plotted.

```python
def myMinMaxScaler(X: np.ndarray):
    X_min = np.min(X)
    X_max = np.max(X)
    s = X_max - X_min
    return (X-X_min)/s

X_scaled = myMinMaxScaler(X)

#Run MLP and Linear Regression with scaled data
mlp_scaled = mlp.fit(X_scaled, y)

y_predict = mlp_scaled.predict(X_scaled)

myXYPlot(X, y, y_predict)
```

As shown in both the models scores and in the plot, the MLP performs much better when X data is scaled.

### Qb) Scikit-learn Pipelines

Next data is scaled with sklearns `sklearn.preprocessing.MinMaxScaler` scaler.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X_MinMaxScaled = scaler.transform(X)

mlp_MinMaxScaled = mlp.fit(X_MinMaxScaled, y)
y_predict = mlp.predict(X_MinMaxScaled)

myXYPlot(X, y, y_predict)
```

Define pipeline with a MinMaxScaler and the mlp model using `sklearn.pipeline.Pipeline`.

```python
from sklearn.pipeline import Pipeline

mlp_pipeline = Pipeline(
    [
        ('scaler',MinMaxScaler()),
        ('mlp',mlp)
    ]
)
```

Fit data using pipeline and plot predictions against data.

```python
mlp_pipe = mlp_pipeline.fit(X,y)

y_pred_pipe = mlp_pipeline.predict(X)

myXYPlot(X, y, y_pred_pipe)
```

### Qc) Outliers and the Min-max Scaler vs. the Standard Scaler

Using a min-max scaler, an outlier (for example single data point much larger/lower than nearest neighbour) will compress the rest of the data after scaling. This is because the scaler is directly calculated from the minimum and maximum values of the data.

Below an outliner, with very large GDP value compared to its Life satisfaction value is introduced into the dataset.

```python
#Introducing an outlier to the dataset
X_with_outlier = X.copy()
X_with_outlier[-1,0] = 150000
```

Model is rerun with the new data, a score is calculated from the model and predticed and actual data is plotted.

```python
mlp_pipe_outlier = mlp_pipeline.fit(X_with_outlier,y)
y_pred_outlier = mlp_pipe_outlier.predict(X_with_outlier)

myXYPlot(X, y, y_pred_outlier)
```

As shown, score from the pipeline (MinMaxScaler and MLP) is very low. From the plot it can be seen how the outlier tilts the preditions towards larger values for the lowest GDP data and vice versa for the largest GDP data.

In the StandardScaler in sklearn (https://scikit-learn.org/dev/modules/generated/sklearn.preprocessing.StandardScaler.html) scaling is performed by subtracting the mean and dividing by the standard deviation. Thus the impact of a single extreme datapoint is mitigated.

### Qd) Modify the MLP Hyperparameters

Below several pipelines are implemented, using different hyperparameters in the MLP.

Lowering the hidden_layer_sizes (number of neurons used) did not impact the prediction capability of the model. In order to investigate the performance of the model the max_iter (number of epochs) was lowerede in conjucntion with the hidden_layer_sizes.
Using only 2 neurons, the lowest number of iterations needed for securing convergence was around 10.000. With 10 neurons around 5.000 epochs were needed.
Thus increasing number of neurons used lowers the number of epochs needed for convergence.

```python
mlp_modified = MLPRegressor(hidden_layer_sizes=(2, ),
                   solver='adam',
                   activation='relu',
                   tol=1E-5,
                   max_iter=10000,
                   verbose=False)

mlp_pipeline_modified = Pipeline(
    [
        ('scaler',MinMaxScaler()),
        ('mlp',mlp_modified)
    ]
)

mlp_pipe_modified = mlp_pipeline_modified.fit(X,y)
y_pred_modified_1 = mlp_pipeline_modified.predict(X)

myXYPlot(X, y, y_pred_modified_1)
```

Changing the activation function to a sigmoid ('logistic') function did not drastically impact the R^2 score. It can be seen however how predictions now follow a sigmoid line.

```python
mlp_sigmoid = MLPRegressor(hidden_layer_sizes=(10, ),
                   solver='adam',
                   activation='logistic',
                   tol=1E-5,
                   max_iter=100000,
                   verbose=False)

mlp_pipeline_sigmoid = Pipeline(
    [
        ('scaler',MinMaxScaler()),
        ('mlp',mlp_sigmoid)
    ]
)

mlp_pipe_sigmoid = mlp_pipeline_sigmoid.fit(X,y)
y_pred_sigmoid = mlp_pipeline_sigmoid.predict(X)

myXYPlot(X, y, y_pred_sigmoid)
```

Changing to a stochastic gradient descent (sgd) with 10 neurons drastically changed the R^2. However lowering the number of neurons to two removed this decrease in the R^2 score.

```python
mlp_sgd = MLPRegressor(hidden_layer_sizes=(2, ),
                   solver='sgd',
                   activation='logistic',
                   tol=1E-5,
                   max_iter=100000,
                   verbose=False)

mlp_pipeline_sgd = Pipeline(
    [
        ('scaler',MinMaxScaler()),
        ('mlp',mlp_sgd)
    ]
)

mlp_pipe_sgd = mlp_pipeline_sgd.fit(X,y)
y_pred_sgd = mlp_pipeline_sgd.predict(X)

myXYPlot(X, y, y_pred_sgd)
```

#### Mathematical foundation of the Linear regression
In linear regression a dataset consisting of a design matrix, X and target values, y_true is used to fit a linear regression model.

$$
\bX =
        \ac{cccc}{
            x_1\pown{1} & x_2\pown{1} & \cdots & x_d\pown{1} \\
            x_1\pown{2} & x_2\pown{2} & \cdots & x_d\pown{2} \\
            \vdots      &             &        & \vdots      \\
            x_1\pownn   & x_2\pownn   & \cdots & x_d\pownn   \\
        } 
$$

$$
\bytrue =
  \ac{c}{
     y\pown{1}\sttrue \\
     y\pown{2}\sttrue \\
     \vdots           \\
     y\pown{n}\sttrue \\
  } 
$$

The linear regression model is a linear equation, consisting of a number of parameters (weights). Here w_0 refers to the bias and all other w are of power 1.

$$
\bw =
    \ac{c}{
         w_0    \\
         w_1    \\
         w_2    \\
         \vdots \\
         w_d    \\
    }
$$

Combining the design matrix with bias terms gives i samples of feature columns, X^(i).

$$
\ar{rl}{
  \ac{c}{1\\\bx\powni} & \mapsto \bx\powni\\
}
$$

The hypothesis model, *h*, refers to a model that can be fitted from data and used for predictions. In a linear equation the hypothesis model makes prediction, y_pred, from a dataset, X, using the weights, w. This is conveniently written in vector notation as shown below. 

$$
\ar{rl}{
  ~~~~~~~~~~~~~~~
  h(\bx\powni;\bw) &= \ypred\powni \\
                   &= \bw^\top \bx\powni ~~~~ (\bx\powni~\textrm{with bias element})\\ 
                   &= w_0  \cdot 1+ w_1 x_1\powni + w_2 x_2\powni + \cdots + w_d x_d\powni & \\
}
$$

In order to train a linear regression model a Loss function is needed function. L^(i) refers to the individual loss, used to describe differences (loss) between single predictions, i, of the linear regression model, y_pred, and the true values from the data, y_true. Again vector notation is a convenient way to write and from this implement the Loss function.

$$
\ar{rl}{
  L\powni &= || \ypred \powni         - \ytrue\powni~ ||_2^2\\
          &= || h(\bx\powni;\bw)      - \ytrue\powni~ ||_2^2\\
          &= || \bw^\top\bx\powni     - \ytrue\powni~ ||_2^2\\
          &= \left( \bw^\top\bx\powni - \ytrue\powni~ \right)^2 ~~~~~ \textrm{(only for 1D output)}
}
$$

The total cost function (or linear regression cost function), J, is proportional to the mean square error, MSE of the loss function. J minimizes the sum of all individual costs.

$$
\ar{rl}{
    \textrm{MSE}(\bX,\bytrue;\bw)  &= \frac{1}{n} \sum_{i=1}^{n} L\powni \\
                                   &= \frac{1}{n} \sum_{i=1}^{n} \left( \bw^\top\bx\powni - y\powni\sttrue \right)^2\\
                                   &= \frac{1}{n} ||\bX \bw - \bytrue||_2^2
}
$$                   



$$
\ar{rl}{
   J &= \frac{1}{2} ||\bX \bw - \bytrue||_2^2\\
     &  \propto \textrm{MSE}
}
$$


By training the model iteratively using the cost function, J, to tweak model weights, $\bw$, optimal weights (the weights that gives the lowest cost function), are found.

$$
  \bw^* = \textrm{argmin}_\bw~J\\
$$

### Implementations

Here a Linear Regresson class, `MyLinReg`, with a fit-predict interface is implemented. The predict interface output a one-dimensional array. The class contains a R^2 score function and a loss function that calculates the gradient based on the MSE. The class constructor contains parameters for the number of iterations (max_iter), learning rate (eta0) and a tolerance (tol).

`MyLinReg` contains three fit methods. One for the closed-form solution (CF), one for a batch gradient descent (GD) and one for a stochastic gradient descent (SGD). For the GD the learning rate is constant and for the SGD an adaptive learning rate is used.

Batch gradient descent and Stochastic gradient descent methods are inspired from HOML 3rd edition, chapter 4 - Gradient Descent.

```python
from sklearn.preprocessing import add_dummy_feature
import numpy as np
```

```python
class MyLinReg():
    def __init__(self, eta0=0.01, max_iter=10, tol=1e-3, n_iter_no_change=5, verbose=True):
        self.eta0 = eta0
        self.max_iter = max_iter
        self.tol = tol
        self.coef_with_intercept_ = None
        self.coef_ = None
        self.intercept_ = None
        self.n_features_in_ = None

    def __str__(self):
        return self.coef_
    
    def lossFunction(self, X, y, theta):
        #Calculates the gradient based on MSE
        return 2/len(X)*X.T@(X@theta-y)
    
    def Weights(self):
        return self.coef_

    #Fit numerically
    def fit_CF(self, X, y):
        X_b = add_dummy_feature(X)
        #Calculate the Normal Equation
        self.coef_with_intercept_ = np.linalg.inv(X_b.T@X_b)@X_b.T@y
        self.intercept_ = self.coef_with_intercept_[0]
        self.coef_ = self.coef_with_intercept_[1:]

        return self.coef_with_intercept_
    
    #Fit with batch gradient descent
    def fit_GD(self, X, y):
        X_b = add_dummy_feature(X)
        #Start with random parameters
        X_dim = np.shape(X_b)
        theta = np.random.randn(X_dim[1],1)
        for epoch in range(self.max_iter):
            #Calculate loss (from gradient)
            loss = self.lossFunction(X_b, y, theta)
            theta = theta - self.eta0*loss
            #Break if loss is below tolerance
            if (np.linalg.norm(loss) < self.tol).all():
                break

        self.coef_with_intercept_ = theta
        self.intercept_ = self.coef_with_intercept_[0]
        self.coef_ = self.coef_with_intercept_[1:]
        return self.coef_with_intercept_

    #Fit stochastic gradient descent
    def fit_SGD(self, X, y):
        X_b = add_dummy_feature(X)
        m = len(X_b)
        #Start with random parameters
        X_dim = np.shape(X_b)
        theta = np.random.randn(X_dim[1],1)

        for epoch in range(self.max_iter):
            for iteration in range(m):
                #Pick random index
                random_index = np.random.randint(m)
                #Pick x_i sample row [w_0, w_1]
                x_i = X_b[random_index: random_index+1]
                #Get y_i of sample
                y_i = y[random_index: random_index+1]
                #Calculate partial derivative of sample
                gradient = self.lossFunction(x_i, y_i, theta)
                theta = theta - self.eta0*gradient
                #Break if gradient is below tolerance
                if (np.linalg.norm(gradient) < self.tol).all():
                    break
                
        self.coef_with_intercept_ = theta
        self.intercept_ = self.coef_with_intercept_[0]
        self.coef_ = self.coef_with_intercept_[1:]
        return self.coef_with_intercept_

    def predict(self, X: np.ndarray):
        X_b = add_dummy_feature(X)
        y_pred = X_b@self.coef_with_intercept_
        return y_pred.flatten()

    def score(self, X, y_true):
        X_b = add_dummy_feature(X)
        y_pred = X_b@self.coef_with_intercept_
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        ss_res = np.sum((y_true - y_pred) ** 2)
        r2_score = 1 - (ss_res / ss_tot)
        return r2_score
```

### Journaling of the Regressors 


#### The Closed form regressor
The method used in the closed form regression fit, `fit_CF(self, X, y)`, uses the Normal Equation (equation below) to calculate the weight parameters (w*) from target (y_true) and feature (X) values in one go. 

$$
  \bw^* ~=~ \left( \bX^\top \bX \right)^{-1}~ \bX^\top \bytrue
$$

The normal equation is implemented as `np.linalg.inv(X_b.T@X_b)@X_b.T@y` where X_b is the feature data including the bias, w_0. .T is a transpose opperation, @ performs matrix multiplication and np.linalg.inv() computes a matrix inversion.

#### The SG regressor

The Batch Gradient Descent (SGD) regressor minimizes the cost function, J, by calculating the gradient (partial differential) of J through the weights.

$$   
  \nabla_\bw~J = \left[ \frac{\partial J}{\partial w_1} ~~~~ \frac{\partial J}{\partial w_2} ~~~~ \ldots  ~~~~ \frac{\partial J}{\partial w_d} \right]^\top
$$

In vector notation this can be written as:

$$
\ar{rl}{
  \nabla_\bw J &= \frac{2}{n} \bX^\top \left( \bX \bw - \bytrue \right)
}
$$

Calculation of the gradient is performed in the `lossFunction()` where gradient is calculated in one step as `2/len(X)*X.T@(X@theta-y)`.

In order to itereratively approach the minimum, at each iteration new weights are calculated by subtracting calculated gradient multiplied by a learning rate (which determines step size). This way steps sizes gets smaller and smaller as the minimum is approached (meaning the algorithm gets slower and slower as a solution is approached).

$$ 
  \bw^{(step~N+1)}~ = \bw^{(step~N)} ~ - \eta \nabla_{\bw} J
$$

This is implemented as a for loop `for epoch in range(self.max_iter)`. In each iteration (epoch) the gradient is calculated and parameters `theta` are updated. To prevent slowing down when approaching small values of the gradient, a tolerance, `tol` is used to stop the for loop at parametized threshold.

During the start of the SG weights are initialized with random values `theta = np.random.randn(X_dim[1],1)`, where `X_dim[1]` are the number of features in X.

The figure below shows the cost, J, of a linear function with weights w_0 and w_1 and how a gradient (blue line) is used to predict the next iterations values of w_0 and w_1, in order to stepwise approach the solution, w*.

#### The SGD regressor

Like the SG regression is performed over a set number of epochs `for epoch in range(self.max_iter)`. In contrast to the SG (where gradient is calculated from the whole dataset) each epoch contains another for loop carrying out a number of iterations `for iteration in range(m)`.

In the Stochastic gradient descent random samples of x^(i) from X (denoted X<sub>SGD</sub>) is drawn for each iteration. This lowers amound of data needed to calculate the gradient of the cost function and in turn the computationel power and time needed. All data (all values of x^(i)) is still used.

$$
\ar{rl}{
  \nabla_\bw J_\text{SGD} &= \frac{1}{n} \bX^\top_\text{SGD} \left( \bX_\text{SGD} \bw - \bytrue \right)
}
$$

This way the gradient does not slowly apprach infinitesimal values and may reach a solution faster than the SG.

In the code a random index, `random_index = np.random.randint(m)`, is used to pick samples of x_i, `x_i = X_b[random_index: random_index+1]`, and y_i `y_i = y[random_index: random_index+1]` from which the gradient is calculated. For now the learning rate, `eta` is a constant.

#### Epochs vs Iterations
When training a machine learning model an epoch refers to one cycle going through the entire training dataset. An iteration refers to an occurence where model parameters are updated. For example, in the implemented SGD method (fit_SGD) the whole dataset, X, is passed through the model a number of times (number of epochs). For each epoch the gradient is calculated for every x^i in the X dataset and the parameters (theta, which contains w_0 and w_1) are updated at each of these iterations. 

Ref: https://www.geeksforgeeks.org/what-is-the-difference-between-epoch-and-iteration-in-training-neural-networks/.

#### Comparison of numerical optimization with Closed-form solution
The numerical (discrete) optimization approximates the closed-form solution by iteration. This gives flexibility and speed in how to arrive at a solution. For example a minimum of the loss function in a dataset can be found through iteration, instead of having to calculate the gradient of a large multidimensionel dataset. The con here is that the solution may have converged on a local minimum instead of the global minimum. Moreover the solution will only be an approximate, not the exact solution and hyperparameters has to be tuned to optimize the model performance.

The Closed-form solution on the other hand, gives an exact solution, if one exists. There is no need for hyperparameter tuning. However if the dataset becomes very large (as it often is in ML)
the task becomes very computationally expensive to arrive at the solution.

Ref: https://machinelearningmastery.com/analytical-vs-numerical-solutions-in-machine-learning/.

To generate a dataset the function `LinearDataGenerator` was implemented. It creates either 1D or 2D dataset of specified size, weights and std_dev.

```python
#Create function for 1D or 2D test data
def LinearDataGenerator(size = 10, w_0 = 1, w_1 = 1, w_2 = 0, max_x = 1,std_dev = 1):
    np.random.seed(42)
    if w_2 == 0:
        X = max_x*np.random.rand(size,1)
        y = w_0+w_1*X+std_dev*np.random.randn(size,1)
    else:
        x1 = max_x*np.random.rand(size,1)
        x2 = max_x*np.random.rand(size,1)
        X = np.column_stack((x1,x2))
        y = w_0+w_1*x1+std_dev*np.random.randn(size,1)+w_2*x2+std_dev*np.random.randn(size,1)
    return X, y
```

###Qa
We need to fit the model using the data generator and  the MLP in the exercise. Then we gotta plot the y_true and y_pred in a graph and extract the network weights, bias coefficients.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# data
np.random.seed(42)
X = np.random.rand(100, 1)  # 100 samples, 1 feature
y_true = 2 * X.squeeze() + 1 + np.random.randn(100) * 0.1  # Linear function with noise

# MLP
mlp = MLPRegressor(activation='tanh',      # Activation function
                   hidden_layer_sizes=[2], # One hidden layer with two neurons
                   alpha=1e-5,             
                   solver='lbfgs',         # Quasi-Newton solver
                   max_iter=10000,         # 10k iterations
                   verbose=True)

mlp.fit(X, y_true)
y_pred = mlp.predict(X)

# Plot true vs predicted values
plt.figure(figsize=(10, 6))
plt.plot(X, y_true, 'o', label='True values')  # True data
plt.plot(X, y_pred, 'r-', label='Predicted values')  # Pred data
plt.title('True vs Predicted values')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Extract and print the weights and biases
weights = mlp.coefs_
biases = mlp.intercepts_

for i, (w, b) in enumerate(zip(weights, biases)):
    if i == 0:
        print(f"Layer {i+1} (Input to Hidden) weights:\n{w}")
        print(f"Layer {i+1} (Input to Hidden) biases:\n{b}")
    else:
        print(f"Layer {i+1} (Hidden to Output) weights:\n{w}")
        print(f"Layer {i+1} (Hidden to Output) biases (Output Bias):\n{b}")

print("Model fitting complete.")
```

So as we can se we get a nice y_pred through the data on the graph, it tends to go in a decent line straight through the middle of the data, with a small curve downwards (which looks to be where there is slightly more data points, meaning this curve is also accurate). Overall the mlp model looks good and we get 7 weights.

1. Hidden Layer Calculations:
Each hidden neuron applies a tanh activation function, with the input x multiplied by its weight and the bias added.

Each hidden neuron applies the tanh activation function to its weighted input (w1*x + b1 for H1, and w2*x + b2 for H2).

#### Qd
We now have to plot the newly made function with np.tanh and x.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh
def tanh(x):
    return np.tanh(x)

# NN 
def neural_network(x, w1, b1, w2, b2, w3, w4, b3):
    # Hidden Layer Activations
    H1 = tanh(w1 * x + b1)
    H2 = tanh(w2 * x + b2)
    
    # Output Layer Calculation
    y_pred = w3 * H1 + w4 * H2 + b3
    
    return y_pred

# Weights and biases from the MLP
w1 = -0.4295
b1 = -0.4097
w2 = 0.9969
b2 = -0.6269
w3 = 0.0340
w4 = 2.1022
b3 = 2.2645

# Generate the input data
np.random.seed(42)
X = np.random.rand(100, 1).squeeze()  # 100 samples, 1 feature
y_true = 2 * X + 1 + np.random.randn(100) * 0.1  # Linear function with noise

# Pred with NN
y_pred = neural_network(X, w1, b1, w2, b2, w3, w4, b3)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(X, y_true, 'o', label='True values')  # True data
plt.plot(X, y_pred, 'r-', label='Predicted values')  # Manually computed predictions
plt.title('True vs Predicted values')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

As we can see the formula perfectly matches the plot made earlier with MLP, so it is all good.

#### Qe
We have to plot half the function as 0.3*tanh(2*x+0.1) and the second part as the remaining bit without its bias. Essentially plotting each section of our function without the bias.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(42)
X = np.linspace(-2, 2, 100)  # 100 points between -2 and 2

# First half of the function
y_math_first_part = 0.3 * np.tanh(2 * X + 0.1)

# Second half of the function
y_math_second_part = -0.3 * np.tanh(5 * X + 3)


# Plot first part
plt.figure(figsize=(10, 6))
plt.plot(X, y_math_first_part, label="First part: $0.3 * tanh(2 * X + 0.1)$", color='blue')
plt.title("First Part of the Function")
plt.xlabel("X")
plt.ylabel("y_math_first_part")
plt.legend()
plt.show()

# Plot second part
plt.figure(figsize=(10, 6))
plt.plot(X, y_math_second_part, label="Second part: $-0.3 * tanh(5 * X + 3)$", color='green')
plt.title("Second Part of the Function")
plt.xlabel("X")
plt.ylabel("y_math_second_part")
plt.legend()
plt.show()
```

First Part: 
𝑦math_first_part=0.3⋅tanh(2⋅X+0.1) represents the first hidden neuron’s contribution. This part looks like a scaled and shifted tanh function.

Second Part: 
𝑦math_second_part =−0.3⋅tanh(5⋅X+3) represents the second hidden neuron’s contribution. The second part will have a slightly different behavior due to the higher scaling factor of 5 in the input, but also being negative compared to the first part.

Both are smilar to a monotonic tanh activation function, as neither of them change direction, they only go positive to negative or from negative to positive.

Combination of Neurons: Each neuron captures a different aspect of the input data. By combining the activations of multiple neurons, the network can approximate non-linear functions. In this case, the two neurons contribute different tanh functions that, when combined, approximate a more complex output. When we add the bias term it allows for further flexibility in shifting the output.

Thus, this network can act as a universal approximator for continuous functions given enough neurons, demonstrating the power of even a "simple" neural networks.

#### Qf
So we have to make it more like a sinc function, by extending the NN with more nodes.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# Generate Sinc Data
def GenerateSincData():
    # A Sinc curve, needs more neurons to capture the oscillations
    X = np.linspace(-3, 3, 1000)
    y = np.sinc(X)
    X = X.reshape(-1, 1)
    return X, y

# Plot Groundtruth Sinc function
X, y_true = GenerateSincData()
plt.plot(X, y_true, "r-", label="True Sinc Function")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Groundtruth Sinc Function")
plt.show()

# MLP
mlp = MLPRegressor(hidden_layer_sizes=(100, 100, 100),  # Three hidden layers with 100 neurons each
                   activation='tanh',                  # Using 'tanh' as the activation function
                   alpha=1e-5,                         
                   solver='adam',                      # Adam solver for optimization
                   max_iter=20000,                     # 20k iterations
                   random_state=42)


mlp.fit(X, y_true)
y_pred = mlp.predict(X)

# Plot predictions vs Groundtruth Sinc
plt.plot(X, y_true, "r-", label="True Sinc Function")
plt.plot(X, y_pred, "b--", label="MLP Approximation")
plt.xlabel("X")
plt.ylabel("y")
plt.title("MLP Approximation of Sinc Function")
plt.legend()
plt.show()
```

We first plot the true sinc function.
Then we plot our approximated mlp with 100 neurons and 3 hidden layers. What we have been able to showcase is, that with a sufficient amount of neurons and layers, the MLP can approximate the oscillations of the sinc function quite well, but not perfectly and confirming the statements written in the last question, about it being able to mimic more complex functions.
